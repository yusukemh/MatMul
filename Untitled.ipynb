```python
# runtime comparison using average running times
fox = np.array([1.5539798, 0.3346818, 0.0590878, 0.0286738, 0.0326626, 0.0590672])
cannon = np.array([1.4635436, 0.2566136, 0.0531246, 0.0204924, 0.0197286, 0.0229716])
outer = np.array([1.4757056, 0.265138, 0.054216, 0.0356182, 0.0460436, 0.0965112])

fig, ax = plt.subplots(1)
x = np.linspace(start=1, stop=400)
num_procs = [1, 4, 16, 64, 100, 400]
line_3 = ax.plot(num_procs, outer, label="outer")
line_1 = ax.plot(num_procs, fox, label="fox")
line_2 = ax.plot(num_procs, cannon, label="cannon")
ax.set_title("Running time", fontsize=20)
ax.legend(fontsize=15)
ax.set_xlabel("# of processors", fontsize=20)
ax.set_ylabel("Runnint time (sec)", fontsize=20)
ax.set_yscale('log')
ax.set_xscale('log')
ax.tick_params(axis='both', which='major', labelsize=15)
fig.savefig("1.png", bbox_inches = 'tight')
```

```python
# parallel speedup using speedup on average running times
fox = np.array([1, 4.6431559, 26.299503, 54.195111, 47.576733, 26.308675])
cannon = np.array([1, 5.7032971, 27.549263, 71.418847, 74.183854, 63.710999])
outer = np.array([1, 5.5658019, 27.219005, 41.431223, 32.050178, 15.290511])

fig, ax = plt.subplots(1)
x = np.linspace(start=1, stop=400)
num_procs = [1, 4, 16, 64, 100, 400]

lie_0 = ax.plot(x, x, label='linear', c='k')
line_3 = ax.plot(num_procs, outer, label='outer', c='b')
line_1 = ax.plot(num_procs, fox, label='fox', c='y')
line_2 = ax.plot(num_procs, cannon, label='cannon', c='r')
ax.set_title("Parallel Speedup", fontsize=20)
ax.legend(fontsize=15)
ax.set_xlabel("# of processors", fontsize=20)
ax.set_ylabel("Parallel Speedup", fontsize=20)
ax.set_yscale('log')
ax.set_xscale('log')
ax.tick_params(axis='both', which='major', labelsize=15)
fig.savefig("2.png", bbox_inches = 'tight')
```

```python
# runtime comparison using average running times
fox = np.array([1.520859, 0.291641, 0.056872, 0.010793, 0.007366, 0.002057])
cannon = np.array([1.559107, 0.267810, 0.049169, 0.010768, 0.007724, 0.002109])
outer = np.array([1.549234, 0.228641, 0.038322, 0.010039, 0.006418, 0.001971])

fig, ax = plt.subplots(1)
x = np.linspace(start=1, stop=400)
num_procs = [1, 4, 16, 64, 100, 400]
line_3 = ax.plot(num_procs, outer, label="outer")
line_1 = ax.plot(num_procs, fox, label="fox")
line_2 = ax.plot(num_procs, cannon, label="cannon")
ax.set_title("Running time", fontsize=20)
ax.legend(fontsize=15)
ax.set_xlabel("# of processors", fontsize=20)
ax.set_ylabel("Runnint time (sec)", fontsize=20)
ax.set_yscale('log')
ax.set_xscale('log')
ax.tick_params(axis='both', which='major', labelsize=15)
fig.savefig("3.png", bbox_inches = 'tight')
```

```python
# parallel speedup using speedup on average running times
fox = np.array([1.520859, 0.291641, 0.056872, 0.010793, 0.007366, 0.002057])
cannon = np.array([1.559107, 0.267810, 0.049169, 0.010768, 0.007724, 0.002109])
outer = np.array([1.549234, 0.228641, 0.038322, 0.010039, 0.006418, 0.001971])

fig, ax = plt.subplots(1)
x = np.linspace(start=1, stop=400)
num_procs = [1, 4, 16, 64, 100, 400]

lie_0 = ax.plot(x, x, label='linear', c='k')
line_3 = ax.plot(num_procs, outer[0]/outer, label='outer', c='b')
line_1 = ax.plot(num_procs, fox[0]/fox, label='fox', c='y')
line_2 = ax.plot(num_procs, cannon[0]/cannon, label='cannon', c='r')
ax.set_title("Parallel Speedup", fontsize=20)
ax.legend(fontsize=15)
ax.set_xlabel("# of processors", fontsize=20)
ax.set_ylabel("Parallel Speedup", fontsize=20)
ax.set_yscale('log')
ax.set_xscale('log')
ax.tick_params(axis='both', which='major', labelsize=15)
fig.savefig("4.png", bbox_inches = 'tight')
```
